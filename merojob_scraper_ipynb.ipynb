```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
def get_jobs(category, pages = 99):
    '''Gets scrapes the job title, company name, location, views and skills'''
    job_list = []
    pages += 1
    for no in range(1,pages):
        # goes through the given number of pages
        url = "https://merojob.com/category/" + category + "/?page="+ str(no)
        r = requests.get(url)
        soup = bs(r.content, 'html.parser')
        jobs = soup.find_all('div',attrs={"class": "card mt-3 hover-shadow"})
        try:
            for x in range(0,6):
                # extracting data
                job_title = jobs[x].find(itemprop="title").get_text(" ", strip=True)
                orgn = orgn = jobs[x].find('h3',attrs={"class": "h6"}).get_text(" ", strip=True)
                location = jobs[x].find(itemprop="address").get_text(" ", strip=True)
                views = jobs[x].find('span',attrs={'class':'text-primary mr-2'}).get_text(" ", strip=True).replace("Views: ","")
                try:
                    desc = jobs[x].find(itemprop="skills").get_text(", ", strip=True)
                except:
                    desc = ("No skills listed")

                # adding data to list

                job_list.append({"Job Title" : job_title,
                "Company" : orgn,
                "Location" : location,
                "Views" : int(views),
                "Skills" : desc})
        except:
            return pd.DataFrame(job_list)#changes the list into a pandas dataframe
    return pd.DataFrame(job_list)#changes the list into a pandas dataframe
```
