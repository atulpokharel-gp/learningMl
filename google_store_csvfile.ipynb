```python
import os
import glob
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import re
from googlesearch import search

query = input()
urls = list(search(query))
social_sites = ["facebook", "twitter", "instagram", "linkedin"]

for url in urls:
    skip_url = False
    for site in social_sites:
        if re.search(site, url, re.IGNORECASE):
            print(f"The URL {url} contains the social media site {site}.")
            skip_url = True
            break
    
    if skip_url:
        continue
    
    try:
        response = requests.get(url, timeout=45)
        response.raise_for_status()
    except (requests.exceptions.HTTPError, requests.exceptions.RequestException):
        print(f"Error occurred while accessing {url}, skipping to next URL.")
        continue
    
    html_content = response.text
    soup = BeautifulSoup(html_content, "html.parser")
    if soup.title is not None:
        title = soup.title.string.strip()
    else:
        title = "No title found"+text[:5]
    text = ""
    video_url = ""
    for element in soup.find_all(["p", "h1", "h2", "h3", "h4", "h5", "h6"]):
        text += element.get_text().strip() + "\n"
    video_elements = soup.find_all("video")
    if len(video_elements) > 0:
        video_url = video_elements[0].get("src")
    filename = 'data'+title[:6]+'.csv'
    with open(filename, "a", newline="") as file:
        writer = csv.writer(file)
        if file.tell() == 0:  # if the file is empty, write the header row
            writer.writerow(["Title", "Text", "URL", "Video URL"])
        writer.writerow([title, text, url, video_url])
    print("Data saved to", filename)

folder_path = '/content/'
all_files = glob.glob(os.path.join(folder_path, '*.csv'))

dfs = []
for f in all_files:
    if os.path.exists(f):  # check if file exists before reading it
        dfs.append(pd.read_csv(f))

if dfs:  # check if there are any dataframes to concatenate
    df = pd.concat(dfs)
    output_file = query+'.csv'
    df.to_csv(output_file, index=False)

    # Delete all the original CSV files
    for f in all_files:
        os.remove(f)
else:
    print("No data found in CSV files.")
```
